import logging
import sys
import os

def setup_custom_logger():
    LOG_DIR = os.getcwd() + '/' + 'logs'
    if not os.path.exists(LOG_DIR):
        os.makedirs(LOG_DIR)
    
    formatter = logging.Formatter(fmt='%(asctime)s %(levelname)-8s %(message)s',
                                  datefmt='%Y-%m-%d %H:%M:%S')
    handler = logging.FileHandler(LOG_DIR+'/log-pengujian-A.txt', mode='w')
    handler.setFormatter(formatter)
    screen_handler = logging.StreamHandler(stream=sys.stdout)
    screen_handler.setFormatter(formatter)
    logger = logging.getLogger()
    logger.setLevel(logging.DEBUG)
    logger.addHandler(handler)
    logger.addHandler(screen_handler)
    return logger

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

model_F = load_model('model_klasifikasiDB-009.model')

#khusus modelDB2-037 (ada 9 kelas)
labels_dict_F={0:'Mengerutkan Dahi',
             1:'Mengangkat Kedua Alis',
             2:'Lirik Kanan',
             3:'Lirik Kiri',
             4:'Normal',
             5:'Lihat Atas',
             6:'Lihat Bawah',
             7:'Lihat Kanan',
             8:'Lihat Kiri',
             9:'Lirik Bawah'}
color_dict_F={0:(236,186,136),
              1:(255,204,255),
              2:(0,0,128),
              3:(211,85,186),
              4:(0,255,0),
              5:(0,155,255),
              6:(255,255,0),
              7:(255,0,0),
              8:(128,128,0), 
              9:(135,184,222)}
```

```python
# deklarasi index dan logger
index = 0
index1 = 0
#logger = setup_custom_logger()
nf=0
h0=0
h1=0
h2=0
h3=0
h4=0
h5=0
h6=0
h7=0
h8=0
h9=0

counterk = 0
countern = 0
counterlka= 0
counterlki= 0
counterno= 0
countera= 0
counterb= 0
counternka= 0
counternki= 0
counterunk= 0
tot=1
```

```python
#dir_path ='C:/Users/MirnaF/Latihan-1/Pengujian/model1/mirna'
dir_path = "E:/DB/alignment-dyo"

for i in os.listdir(dir_path):
    print (i)
    img = cv2.imread(dir_path+'//'+i)
    
    #bila 
    if img is None:
        print("Gambar Tidak Dapat Diakses")
        index1+=1
    
    else:
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

        dets = detector(img)

        num_faces = len(dets)
        #print("banyaknya wajah:",num_faces)

        if num_faces == 0:
            print("Sorry, there were no faces found")
            nf += 1
            #logger.debug('Face Not Found')
            cv2.putText(img, "NO FACE", (200,240), cv2.FONT_HERSHEY_PLAIN, 4, (0,0,255), 5)
        else:
            # Find the 5 face landmarks we need to do the alignment.
            faces = dlib.full_object_detections()
            try:
                for detection in dets:
                    # predictor tuh sp asalnya
                    faces.append(predictor(img, detection))

                images = dlib.get_face_chips(img, faces, size=320)
                
                for image in images:

                    #konversi ke skala abu-abu
                    kerangkaAbu = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

                resized_F=cv2.resize(kerangkaAbu,(150,150))
                normalized_F=resized_F/255.0
                reshaped_F=np.reshape(normalized_F,(1,150,150,1))
                result_F=model_F.predict(reshaped_F)

                label_F=np.argmax(result_F,axis=1)[0]
                #cv2.imshow('REGION FACE',image)
                print(labels_dict_F[label_F])
                #logger.info(labels_dict_F[label_F])
                
            #untuk tampilan saja
                if (label_F != 0) & (label_F != 1):        
                    cv2.rectangle(img,(10,400),(200,450),color_dict_F[label_F],-1)
                    cv2.putText(img, labels_dict_F[label_F], (20,430),cv2.FONT_HERSHEY_SIMPLEX,0.8, (255,255,255),2)

                else: 
                #ukuran latar untuk mengangkat kedua alis
                    cv2.rectangle(img,(10,400),(350,450),color_dict_F[label_F],-1)
                    cv2.putText(img, labels_dict_F[label_F], (10,430),cv2.FONT_HERSHEY_SIMPLEX,0.8, (255,255,255),2)

                print(".......................")

                if label_F == 0:
                    h0 += 1
                    print("alis kerung")
                    plt.imshow(img)
                    plt.axis('off')   #untuk menghilangkan axis nya
                    fig1 = plt.gcf()

                    path1 = 'D:/DATASET/KLASIFIKASI/kerung'
                    fig1.savefig((os.path.join(path1 , f'v2-krg({counterk}).jpg')), bbox_inches='tight',transparent=True, pad_inches=0)

                    print("Saved!")
                    counterk = counterk+1
                    
                elif label_F == 1:
                    h1 += 1
                    print("alis naik")
                    plt.imshow(img)
                    plt.axis('off')   #untuk menghilangkan axis nya
                    fig1 = plt.gcf()

                    path1 = 'D:/DATASET/KLASIFIKASI/naik'
                    fig1.savefig((os.path.join(path1 , f'v2-naik({countern}).jpg')), bbox_inches='tight',transparent=True, pad_inches=0)

                    print("Saved!")
                    countern = countern+1
                    
                elif label_F == 2:
                    h2 += 1
                    print("lirik kanan")
                    plt.imshow(img)
                    plt.axis('off')   #untuk menghilangkan axis nya
                    fig1 = plt.gcf()

                    path1 = 'D:/DATASET/KLASIFIKASI/kanan'
                    fig1.savefig((os.path.join(path1 , f'v2-kan({counterlka}).jpg')), bbox_inches='tight',transparent=True, pad_inches=0)

                    print("Saved!")
                    counterlka = counterlka+1
                    
                elif label_F == 3:
                    h3 += 1
                    print("lirik kiri")
        
                    plt.imshow(img)
                    plt.axis('off')   #untuk menghilangkan axis nya
                    fig1 = plt.gcf()

                    path1 = 'D:/DATASET/KLASIFIKASI/kiri'
                    fig1.savefig((os.path.join(path1 , f'v2-kir({counterlki}).jpg')), bbox_inches='tight',transparent=True, pad_inches=0)

                    print("Saved!")
                    counterlki = counterlki+1
                elif label_F == 4:
                    h4 += 1
                    print("normal")
        
                    plt.imshow(img)
                    plt.axis('off')   #untuk menghilangkan axis nya
                    fig1 = plt.gcf()

                    path1 = 'D:/DATASET/KLASIFIKASI/normal'
                    fig1.savefig((os.path.join(path1 , f'v2-norm({counterno}).jpg')), bbox_inches='tight',transparent=True, pad_inches=0)

                    print("Saved!")
                    counterno = counterno+1
                elif label_F == 5:
                    h5 += 1
                    print("ngadep atas")
        
                    plt.imshow(img)
                    plt.axis('off')   #untuk menghilangkan axis nya
                    fig1 = plt.gcf()

                    path1 = 'D:/DATASET/KLASIFIKASI/ngadep-atas'
                    fig1.savefig((os.path.join(path1 , f'v2-nga({countera}).jpg')), bbox_inches='tight',transparent=True, pad_inches=0)

                    print("Saved!")
                    countera = countera+1
                elif label_F == 6:
                    h6 += 1
                    print("ngadep bawah")

                    plt.imshow(img)
                    plt.axis('off')   #untuk menghilangkan axis nya
                    fig1 = plt.gcf()

                    path1 = 'D:/DATASET/KLASIFIKASI/ngadep-bawah'
                    fig1.savefig((os.path.join(path1 , f'v2-ngba({counterb}).jpg')), bbox_inches='tight',transparent=True, pad_inches=0)

                    print("Saved!")
                    counterb = counterb+1
                elif label_F == 7:
                    h7 += 1
                    print("ngadep kanan")
        
                    plt.imshow(img)
                    plt.axis('off')   #untuk menghilangkan axis nya
                    fig1 = plt.gcf()

                    path1 = 'D:/DATASET/KLASIFIKASI/ngadep-kanan'
                    fig1.savefig((os.path.join(path1 , f'v2-ngka({counternka}).jpg')), bbox_inches='tight',transparent=True, pad_inches=0)

                    print("Saved!")
                    counternka = counternka+1

                elif label_F == 8:
                    h8 += 1
                    print("ngadep kiri")

                    plt.imshow(img)
                    plt.axis('off')   #untuk menghilangkan axis nya
                    fig1 = plt.gcf()

                    path1 = 'D:/DATASET/KLASIFIKASI/ngadep-kiri'
                    fig1.savefig((os.path.join(path1 , f'v2-ngkir({counternki}).jpg')), bbox_inches='tight',transparent=True, pad_inches=0)

                    print("Saved!")
                    counternki = counternki+1
                elif label_F == 9:
                    h9 += 1
                    print("Unknow")

                    plt.imshow(img)
                    plt.axis('off')   #untuk menghilangkan axis nya
                    fig1 = plt.gcf()

                    path1 = 'D:/DATASET/KLASIFIKASI/unknown'
                    fig1.savefig((os.path.join(path1 , f'v2-unknow({counternki}).jpg')), bbox_inches='tight',transparent=True, pad_inches=0)

                    print("Saved!")
                    counterunk = counterunk+1

                index+=1
                tot=tot+1

            except cv2.error as e:
                print (e)
                index1 += 1
        
        print (labels_dict_F[0] + " : " + str(h0))
        print (labels_dict_F[1] + " : " + str(h1))
        print (labels_dict_F[2] + " : " + str(h2))
        print (labels_dict_F[3] + " : " + str(h3))
        print (labels_dict_F[4] + " : " + str(h4))
        print (labels_dict_F[5] + " : " + str(h5))
        print (labels_dict_F[6] + " : " + str(h6))
        print (labels_dict_F[7] + " : " + str(h7))
        print (labels_dict_F[8] + " : " + str(h8))
        print (labels_dict_F[9] + " : " + str(h9))
        print ("No Face : ", nf)
        print ("Tak Tebaca : ", index1)
        print ("Total : ", index)
        print ("-----------------------------------------------------------")

        scale_percent = 50 # percent of original size
        width = int(img.shape[1] * scale_percent / 100)
        height = int(img.shape[0] * scale_percent / 100)
        dim = (width, height)
        
        #img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA) #resize image
        #cv2.imshow('LIVE',img)
        #cv2.waitKey(0)
        
    #cv2.destroyAllWindows()
```
