```python
import matplotlib.pyplot as plt
import cv2,os

#data_path='dataset-bymir/by_mir'
data_path='C:/Users/MirnaF/Latihan-1/DB-FIX/DB-classifier_v3'
# 'C:/Users/MirnaF/Latihan-1/DB-FIX/P12/normal' --> square --> dataDB2 dan targetDB2
# 'C:/Users/MirnaF/Latihan-1/DB-FIX/P12/wajah' --> wajah square --> dataWS dan targetWS
# 'C:/Users/MirnaF/Latihan-1/DB-FIX/P12/ROI/normalization' --> mata alis ROI --> dataRN dan targetRN
categories=os.listdir(data_path)
labels=[i for i in range(len(categories))]

label_dict=dict(zip(categories,labels))

print(label_dict)
print(categories)
print(labels)
```

```python
img_size=150

#tinggi=108
#lebar=241
data=[]
target=[]


for category in categories:
    folder_path=os.path.join(data_path,category)
    img_names=os.listdir(folder_path)   #alamat si FOLDER
    print (folder_path)
    print ('--------------------------------------')
    print (img_names)
    print ('**************************************')
    
    for img_name in img_names:
        img_path=os.path.join(folder_path,img_name)
        img=cv2.imread(img_path)

        #cv2.imshow("asli", img)
        #cv2.waitKey(0)
        #cv2.destroyAllWindows()

        try:
            gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)           
            #Coverting the image into gray scale
            
            resized=cv2.resize(gray,(img_size,img_size)) #untuk wajah, kan square
            #resized=cv2.resize(gray,(lebar,tinggi)) #untuk ROI dia kan landscape
            #resizing the gray scale into 100x100, since we need a fixed common size for all the images in the dataset
            
            data.append(resized)
            target.append(label_dict[category])
            #appending the image and the label(categorized) into the list (dataset)
   
            #coba memunculkan
#            cv2.imshow("Gray", gray)
#            cv2.imshow("Resized image", resized)
#            cv2.waitKey(0)
#            cv2.destroyAllWindows()
  
        except Exception as e:
            print('Exception:',e)
            #if any exception rasied, the exception will be printed here. And pass to the next image
```

```python
import numpy as np
data=np.array(data)/255.0
print(data.shape)
```

```python
#mereshape image yg ada dalam data (berupa array)

data=np.reshape(data,(data.shape[0],img_size,img_size,1)) #untuk yg wajah (square)
#data=np.reshape(data,(data.shape[0],lebar,tinggi,1)) #untuk yg ROI (landscape)

target=np.array(target)

from keras.utils import np_utils

new_target=np_utils.to_categorical(target)
```

```python
np.save('data-klasifikasiDB',data)
np.save('target-klasifikasiDB',new_target)
```
